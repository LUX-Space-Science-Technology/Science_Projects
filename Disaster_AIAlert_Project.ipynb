# General Operations

```python
class ErrorModule(object):
  "Error Module"
  def __init__(self,initialerror:ERRORTYPE,initialmessage:str)->CLASSINIT:
    self.errors = initialerror
    self.message = initialmessage
  def __str__(self)->str:
    return "Error Module Creation - Script Initial Process"
  def __call__(self)->NULLPROCESS:
    return self.errors(self.message)
  def __getstate__(self)->NULLPROCESS:
    raise TypeError(generalTypeError)
  def __repr__(self)->DOCUMENTATION:
    return ErrorModule.__doc__
  def RunRaise(self)->PROCESSTYPE:
    raise self.errors(self.message)
  def RunNone(self)->None:
    return None
```

```python
class SaveFileLocal(object):
  def __init__(self,savetopath:str)->CLASSINIT:
    self.dir = savetopath
    self.ops = None
  def __str__(self)->str:
    return "Saving File - Script Initial Process"
  def __call__(self)->NULLPROCESS:
    return None
  def __getstate__(self)->NULLPROCESS:
    ErrorModule(TypeError,generalTypeError).RunRaise()
  def __repr__(self)->DOCUMENTATION:
    return SaveFileLocal.__doc__
  def WriteOn(self,txtinitial:str)->PROCESSTYPE:
    if isinstance(txtinitial,str):
      self.ops.write(txtinitial)
    else:
      self.ops.write(str(txtinitial))
  def __enter__(self)->CLASSINIT:
    self.ops = open(self.dir,"w",errors="replace")
    return self
  def __exit__(self,ety:NULLPROCESS,evl:NULLPROCESS,etb:NULLPROCESS)->PROCESSTYPE:
    self.ops.close()
```

```python
class GetMeteoData(object):
  def __init__(self)->CLASSINIT:
    self.data = meteosource
    self.feed = None
    self.multidatalist = []
  def __str__(self)->str:
    return "Meteoalarm Data Parsing - Process"
  def __call__(self)->NULLPROCESS:
    return None
  def __getstate__(self)->NULLPROCESS:
    ErrorModule(TypeError,generalTypeError).RunRaise()
  def __repr__(self)->DOCUMENTATION:
    return GetMeteoData.__doc__
  def ReadParser(self,urlinit:URLTYPE)->DATATYPE:
    try:
      self.feed = feedparser.parse(urlinit)
      return self.feed
    except:
      pass
  def GetDocument(self)->PROCESSTYPE:
    for xname,xsite in self.data.items():
      feed_ = self.ReadParser(xsite)
      if feed_ is not None:
        entries_ = feed_["entries"]
        if len(entries_) > 0 and isinstance(entries_,list):
          print(f"[PROCESS - CORRECT] FOR {xname}")
          for index in range(len(entries_)):
            data_ = entries_[index]
            if data_["cap_message_type"].lower() == "alert":
              try:
                area_ = data_["cap_areadesc"]
                country_ = xname
                time_ = data_["cap_sent"]
                event_ = data_["cap_event"]
                status_ = data_["cap_status"]
                self.multidatalist.append([str(area_),
                                          str(country_),
                                          str(time_),
                                          str(event_),
                                          str(status_)])
              except:
                pass
            else:
              pass
        else:
          pass
        print(f"[SAVED] FOR {xname}\n"+"####"*5+"\n\n")
      else:
        pass
    if len(self.multidatalist) > 0:
      return self.multidatalist
    else:
      return False
```

```python
def GetWikiDocument(topic:str,node_list:list,limit_search:int=30,**load_kwargs:typing.Any)->DATATYPE:
  try:
    target_content = ""
    engine_ = wpp.search(str(topic),results=limit_search)
    if (len(engine_) > 0) and (engine_ is not None):
      for t_ in engine_:
        try:
          initial_content = str(wpp.page(str(t_),**load_kwargs).content)
          node_list.append(Document(text=initial_content))
          target_content += initial_content + "\n"
        except wpp.exceptions.DisambiguationError:
          pass
        except Exception as err:
          print(str(err))
          pass
      with SaveFileLocal(wikidatapath) as sfile:
        sfile.WriteOn(target_content)
    else:
      pass
  except Exception as err:
    print(str(err))
    pass
```

## Model Structure

```python
class MODELTYPE:
  pass
class NODETYPE:
  pass
class SERVICETYPE:
  pass
class EVALUATORTYPE:
  pass
```

```python
class QueryEngine(object):
  def __init__(self)->CLASSINIT:
    self.wikidir = wikidatapath
    self.meteodir = meteodatapath
  def __str__(self)->str:
    return "Query Engine Model - Model Process"
  def __call__(self)->NULLPROCESS:
    return None
  def __getstate__(self)->NULLPROCESS:
    ErrorModule(TypeError,generalTypeError).RunRaise()
  def __repr__(self)->DOCUMENTATION:
    return QueryEngine.__doc__
  def ReturnDebugHandler(self,trace_on:bool=True)->PROCESSTYPE:
    db_ = LlamaDebugHandler(print_trace_on_end=trace_on)
    manager_ = CallbackManager([db_])
    return manager_,db_
  def LoadDocument(self,pathinitial:str)->DATATYPE:
    if os.path.exists(pathinitial):
      doc_ = SimpleDirectoryReader(input_files=[pathinitial]).load_data()
      return doc_
    else:
      return False
  def DefineEngines(self)->MODELTYPE:
    doc_wiki = self.LoadDocument(self.wikidir)
    if doc_wiki is not False:
      vector_wiki = VectorStoreIndex.from_documents(doc_wiki)
      engine_wiki = vector_wiki.as_query_engine(similarity_top_k=1)
    else:
      engine_wiki = False
    doc_meteo = self.LoadDocument(self.meteodir)
    if doc_meteo is not False:
      vector_meteo = VectorStoreIndex.from_documents(doc_meteo)
      engine_meteo = vector_meteo.as_query_engine(similarity_top_k=1)
    else:
      engine_meteo = False
    return engine_wiki,engine_meteo
  def DefineTools(self)->PROCESSTYPE:
    wiki_,meteo_ = self.DefineEngines()
    if (wiki_ is not False) and (meteo_ is not False):
      tool_ = [QueryEngineTool(query_engine=wiki_,
                              metadata=ToolMetadata(name="Wikipedia Engine Model",
                                                    description="Based on Wikipedia Searching Data")),
              QueryEngineTool(query_engine=meteo_,
                              metadata=ToolMetadata(name="Meteoalarm Engine Model",
                                                    description="Based on Meteoalarm System Data"))]
    else:
      tool_ = False
    return tool_
  def LoadSystem(self)->MODELTYPE:
    tool_ = self.DefineTools()
    if tool_ is not False:
      manager_,db_ = self.ReturnDebugHandler()
      service_ = ServiceContext.from_defaults(callback_manager=manager_)
      engine_ = SubQuestionQueryEngine.from_defaults(query_engine_tools=tool_,verbose=True,use_async=False,service_context=service_)
    else:
      engine_ = None
    return engine_,db_
```

```python
class EmbedEngine(object):
  def __init__(self)->CLASSINIT:
    self.document = SimpleDirectoryReader(input_files=[meteodatapath]).load_data()
    self.modelname="gpt-3.5-turbo-16k"
    self.maxtoken = 170
  def __str__(self)->str:
    return "Embedding Model Structure - Model Process"
  def __call__(self)->NULLPROCESS:
    return None
  def __getstate__(self)->NULLPROCESS:
    ErrorModule(TypeError,generalTypeError).RunRaise()
  def __repr__(self)->DOCUMENTATION:
    return EmbedEngine.__doc__
  def GetBaseModel(self,temperature:float or int,presence:float or int,frequency:float or int)->MODELTYPE:
    return OpenAI(model=self.modelname,
                  temperature=temperature,
                  presence_penalty=presence,
                  frequency_penalty=frequency,
                  max_tokens=self.maxtoken)
  def GetService(self,predictor:MODELTYPE)->SERVICETYPE:
    return ServiceContext.from_defaults(llm=predictor)
  def GetVectorModel(self,nodes:NODETYPE,service:SERVICETYPE,pathtosave:str)->PROCESSTYPE:
    smp = VectorStoreIndex.from_documents(nodes,service_context=service)
    if os.path.exists(pathtosave):
      isFull = controlLength(pathtosave)
      if isFull:
        store_ = StorageContext.from_defaults(persist_dir=pathtosave)
        smp = load_index_from_storage(store_)
        return smp
      else:
        smp.storage_context.persist(persist_dir=pathtosave)
        return smp
    else:
      return False
  def LoadSystem(self,pathtosave:str,temperature:float or int,presence:float or int,frequency:float or int)->MODELTYPE:
    prediction_model = self.GetBaseModel(temperature=temperature,presence=presence,frequency=frequency)
    service_ = self.GetService(predictor=prediction_model)
    base_model = self.GetVectorModel(nodes=self.document,service=service_,pathtosave=pathtosave)
    if base_model is not False:
      return base_model
    else:
      return False
```

```python
class RetryEngine(object):
  def __init__(self)->CLASSINIT:
    self.document = SimpleDirectoryReader(input_files=[meteodatapath,wikidatapath]).load_data()
  def __str__(self)->str:
    return "Retry Model Structure - Model Process"
  def __call__(self)->NULLPROCESS:
    return None
  def __getstate__(self)->NULLPROCESS:
    ErrorModule(TypeError,generalTypeError).RunRaise()
  def __repr__(self)->DOCUMENTATION:
    return RetryEngine.__doc__
  def GetEvaluator(self)->EVALUATORTYPE:
    return QueryResponseEvaluator()
  def GetVectorModel(self)->MODELTYPE:
    return VectorStoreIndex.from_documents(self.document)
  def GetEngine(self)->MODELTYPE:
    vector_ = self.GetVectorModel()
    return vector_.as_query_engine()
  def GetRetry(self)->MODELTYPE:
    vector_engine = self.GetEngine()
    evaluator_ = self.GetEvaluator()
    retry_ = RetryQueryEngine(vector_engine,evaluator_)
    return retry_
  def GetGuideline(self)->PROCESSTYPE:
    guideline_eval = GuidelineEvaluator(
    guidelines=DEFAULT_GUIDELINES + "\nYour answer should be simple and short.\n"
    "State scientific and clear results.\n")
    return guideline_eval
  def ResponseControl(self,queryinitial:str)->PROCESSTYPE:
    guid_ = self.GetGuideline()
    engine_ = self.GetEngine()
    response = engine_.query(queryinitial)
    print(f"Vector Response: {response}")
    type_ = response if isinstance(response,Response) else response.get_response()
    eval_ = guid_.evaluate_response(queryinitial,type_)
    print(f"Prompt Evaluation: {eval_.feedback}")
    feed_transform = FeedbackQueryTransformation(resynthesize_query=True)
    query_transform = feed_transform.run(queryinitial,{"evaluation":eval_})
    print(f"Query Transformation: {query_transform.query_str}")
  def GetRetrySource(self)->MODELTYPE:
    eval_ = self.GetEvaluator()
    vector_ = self.GetEngine()
    retry_source = RetrySourceQueryEngine(vector_,eval_)
    return retry_source
  def LoadSystem(self,queryinitial:str)->MODELTYPE:
    vector_ = self.GetEngine()
    guide_ = self.GetGuideline()
    self.ResponseControl(queryinitial)
    retry_engine_with_guide = RetryGuidelineQueryEngine(vector_,guide_,resynthesize_query=True)
    response_ = retry_engine_with_guide.query(queryinitial)
    return response_
```

```python
def ConditionalEngine(basemodel:MODELTYPE,promptinitial:str)->MODELTYPE:
  model_ = basemodel.as_query_engine()
  chat_ = CondenseQuestionChatEngine.from_defaults(query_engine=model_,
                                                   condense_question_prompt=promptinitial,
                                                   verbose=True)
  return chat_
```
